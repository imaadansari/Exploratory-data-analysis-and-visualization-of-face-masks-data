```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from textblob import TextBlob
```

```python
df=pd.read_excel("new.xlsx")
prd_df=pd.read_excel("products.xlsx")
prd_df.set_index('product_id')
```

Making a common column of translation and english reviews.

```python
#Making a common column of translation and english reviews.
df["all_reviews"]=df["translation.reviewText"]
for i in range(len(df)):
  if type(df["all_reviews"][i])==float:
    df["all_reviews"][i]=df["reviewText"][i]
```

```python
df
```

```python
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')
```

Data Cleaning

```python
#Data Cleaning
df['word_list']=df['all_reviews']
df['word_list']=df['word_list'].apply(lambda x: [item for item in x.replace(".","").replace("!","").replace(",","").strip().lower().split(" ") if item not in stop])
```

Sentiment Analysis

```python
#Sentiment Analysis
df['polarity_score']=df['all_reviews']
df['polarity_score']=df['polarity_score'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df.to_excel("completed2.xlsx")
```

```python
d={}
def count_words(words):
  for word in words:
          if word in d:
              d[word] = d[word] + 1
          else:
              d[word] = 1

for i in df['word_list']:
  count_words(i)
```

```python
#Sorting keywords based on their occurances.
print(sorted(d.items(), key=lambda x:x[1],reverse=True))
```
